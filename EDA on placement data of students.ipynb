#### fillling null values with 0

```python
df = df.fillna('0')
```

#### converting p_status into 0 and 1

```python
p_status = []
for i in df['status']:
    if i == 'Placed':
        p_status.append(int(1))
    else:
        p_status.append(int(0))
```

```python
df['p_status'] = p_status
```

## 4)statistical test

```python
import scipy.stats as stats
import statsmodels.api as sm

print("Descriptive Statistics:")
print(df.describe())
print('\n')

# t-test
group1_data = df[df['hsc_s'] == 'Science']['p_status']
group2_data = df[df['degree_t'] == 'Sci&Tech']['p_status']
t_statistic, p_value = stats.ttest_ind(group1_data, group2_data)
print("T-test:")
print("T-statistic:", t_statistic)
print("P-value:", p_value)
print('\n')


# Chi-square test
contingency_table = pd.crosstab(df['hsc_s'], df['degree_t'])
chi2_statistic, p_value, _, _ = stats.chi2_contingency(contingency_table)
print("Chi-square test:")
print("Chi-square statistic:", chi2_statistic)
print("P-value:", p_value)
print('\n')


# Correlation
correlation_matrix = df.corr()
print("Correlation:")
print(correlation_matrix)
```
